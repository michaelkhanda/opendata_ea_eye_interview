```python
# Clean and structure the data

# Extract referee information
referees = data.get("referees", [])

# Extract match date and time
date_time = data.get("date_time", "")

# Extract home and away team information
home_team = data.get("home_team", {})
away_team = data.get("away_team", {})

# Extract pitch length
pitch_length = data.get("pitch_length", 0)

# Extract player information
players = data.get("players", [])

# Clean player data
cleaned_players = []
for player in players:
    cleaned_player = {
        "first_name": player.get("first_name", ""),
        "last_name": player.get("last_name", ""),
        "number": player.get("number", ""),
        "birthday": player.get("birthday", ""),
        "player_role": player["player_role"]["name"] if "player_role" in player else "",
    }
    cleaned_players.append(cleaned_player)

# Now you have the cleaned data in the 'referees', 'date_time', 'home_team', 'away_team', 'pitch_length', and 'cleaned_players' variables, which you can use for further processing or analysis.

# Example: Printing the cleaned data
print("Referees:", referees)
print("Date and Time:", date_time)
print("Home Team:", home_team)
print("Away Team:", away_team)
print("Pitch Length:", pitch_length)
print("Players:", cleaned_players)
```
